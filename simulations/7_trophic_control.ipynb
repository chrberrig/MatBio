```python
# Import stuff: 
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

from sympy import Symbol
from sympy.solvers import solve

# This is for reasonable fontsize universally defined:
fs_label = 20
parameters = {                
                'figure.titlesize': fs_label+6,
                'axes.labelsize': fs_label,
                'axes.titlesize': fs_label+4,
                'xtick.labelsize': fs_label,
                'ytick.labelsize': fs_label, 
                'legend.fontsize': fs_label, 
                'lines.linewidth': 6,
            }
plt.rcParams.update(parameters)
```

```python
# Question 1.

t = np.linspace(0, 1000, 10000) # we run it for so long bc we want to be sure the system reaches equilibrium

r = 1            # Growth rate of producers. Dimension: 1/t, Unit: 1/year
K = 15           # Carrying capacity of producers. Dimension: weight/area, Unit: kg/sqm
bc = 1/15        # Clearance rate of consumers. Dimension: area/time/consumer, Unit: sqm/year/#consumers # was 1/50
dc = 1/2         # Death rate of consumers. Dimension: 1/t, Unit: 1/year # was 1/5
epsc = 0.8       # Efficiency of consumption. Dimension: consumers/producers, Unit: #consumers/kg

# Christian's values:
# r = 1
# K = 20
# bc = 1/50
# dc = 1/5
# epsc = 0.6


def deriv(state, t, *params):
    N, C = state
    r, K, bc, epsc, dc = params
    dNdt = r*(1-N/K)*N - bc*C*N
    dCdt = epsc*bc*C*N - dc*C
    return np.array([dNdt, dCdt])
```

```python
N_star, C_star = K, 0.01

K_crit = dc/(epsc*bc)
print(K_crit, (r/bc)*(1-K_crit/K))

params = (r, K, bc, epsc, dc)
state_init = np.array([N_star, C_star])

ns = odeint(deriv, state_init, t, args=params)
N, C = ns.T
```

```python
# Checking the results of the runs (not for the ppt)
fig, ax = plt.subplots(figsize=(16,8))
ax.plot(t, N, label="N")
ax.plot(t, C, label="C")
#ax.set_yscale("log")
ax.legend()

# !NB the units of N and C are not the same!
```

```python
# Question 2: Making bifurcation diagram

N_star, C_star = K, 0.02 # initial N and C are very close to equilibrium, so that the computations are faster

k_list = np.arange(0.1, 75, 0.1) # make array of K values
eq_list = [] # make list to place the values of N and C when they reach equilibrium
for k in k_list:
    params = (r, k, bc, epsc, dc)
    state_init = np.array([N_star, C_star]) 

    ns = odeint(deriv, state_init, t, args=params) 
    N, C = ns.T
    N_star, C_star = N[-1], C[-1] # assuming that all runs reached equilibrium, we save the last value of N and C as the equilibria
    eq_list.append([N_star, C_star])
    N_star += 0.1 # Then we pertrub a bit, because if we start from the exact equilibrium, the simulation gets stuck
    C_star += 0.1
    
N_eq, C_eq = np.array(eq_list).T # Put the list of equilibria in an array, and transpose 
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.plot(k_list, N_eq, label="$N^{*}$", color = 'mediumseagreen')
ax.plot(k_list, C_eq, label="$C^{*}$", color = 'mediumvioletred')
ax.set_xlabel("Carrying capacity K, units $\\frac{kg}{m^{2}}$")
ax.set_ylabel("Producer (N) and Consumer (C) density")
ax.axvline(K_crit, alpha=0.5, linestyle="--", color='mediumvioletred', label="$K_{crit} = \\frac{d_{c}}{\\varepsilon_{c}*b_{c}}$")
ax.legend()
ax.grid(alpha = 0.5)
```

```python
# Question 3: Add a predator
t = np.linspace(0, 1000, 10000)

r = 1            # Growth rate of producers. Dimension: 1/t, Unit: 1/year
K = 15           # Carrying capacity of producers. Dimension: weight/area, Unit: kg/sqm
bc = 1/15        # Clearance rate of consumers. Dimension: area/time/consumer, Unit: sqm/year/#consumers
dc = 1/2         # Death rate of consumers. Dimension: 1/t, Unit: 1/year
epsc = 0.8       # Efficiency of consumption. Dimension: consumers/producers, Unit: #consumers/kg

bp = 1/10        # Clearance rate of predators. Dimension: area/time/predator, Unit: sqm/year/#predators
dp = 1/6         # Death rate of consumers. Dimension: 1/t, Unit: 1/year
epsp = 0.7       # Efficiency of consumption. Dimension: predators/consumers, Unit: #predators/#consumers

# Original values from Christian
# r = 1
# K = 20
# bc = 1/50
# dc = 1/5
# epsc = 0.6
# bp = 1/50
# dp = 1/5
# epsp = 0.4

def deriv_predator(state, t, *params):
    N, C, P = state
    r, K, bc, epsc, dc, bp, epsp, dp = params
    dNdt = r*(1-N/K)*N - bc*C*N
    dCdt = epsc*bc*C*N - dc*C - bp*C*P
    dPdt = epsp*bp*C*P - dp*P
    return np.array([dNdt, dCdt, dPdt])
```

```python
C_star = dp/(epsp*bp) # from \dot{P} = 0
N_star = K*(1-(bc/r)*C_star) # from \dot{N} = 0
P_star = (epsc*bc*N_star - dc)/bp # from \dot{C} = 0

print(N_star)
print(C_star)
print(P_star)


offset = 2.
params = (r, K, bc, epsc, dc, bp, epsp, dp)
state_init = np.array([N_star+offset, C_star-offset, P_star+offset])

ns = odeint(deriv_predator, state_init, t, args=params)
N, C, P = ns.T
```

```python
fig, ax = plt.subplots(figsize=(16,8))
## plotting subpopulations:
ax.plot(t, N, label="N")
ax.plot(t, C, label="C")
ax.plot(t, P, label="P")
## plotting population of entire system:
# ax.plot(t, N+C+P, label="P")
## plotting expected equilibria:
eq_line_style = {"alpha":0.3, "linestyle":"--", "color":"magenta"}
ax.axhline(N_star, **eq_line_style, label="Unstable equilibria")
ax.axhline(C_star, **eq_line_style)
ax.axhline(P_star, **eq_line_style)

ax.legend()

# !NB the units of N and C are not the same!
```

```python
fig, axs = plt.subplots(1, 3, figsize=(16,8))
## plotting subpopulations:
axs[0].plot(N, C, label="N vs C")
axs[0].set_xlabel("N")
axs[0].set_ylabel("C")
axs[1].plot(C, P, label="C vs P")
axs[0].set_xlabel("C")
axs[0].set_ylabel("P")
axs[2].plot(P, N, label="P vs N")
axs[0].set_xlabel("P")
axs[0].set_ylabel("N")
for ax in axs:
    ax.legend()
```

```python
N_star, C_star, P_star = K, 0.01, 0.01

k_list = np.arange(0.1, 40, 0.1)
eq_list = []
for k in k_list:
    params = (r, k, bc, epsc, dc, bp, epsp, dp)
    state_init = np.array([N_star, C_star, P_star])

    ns = odeint(deriv_predator, state_init, t, args=params)
    N, C, P = ns.T
    N_star, C_star, P_star = N[-1], C[-1], P[-1]
    eq_list.append([N_star, C_star, P_star])
    N_star += 0.1
    C_star += 0.1
    P_star += 0.1
    
N_eq, C_eq, P_eq = np.array(eq_list).T
```

```python
r_krit = (bc*dp)/(epsp*bp)

K_crit1 = dc/(epsc*bc)
K_crit2 = (dc/(epsc*bc))/(1-C_star*bc/r)

C_crit = dp/(epsp*bp)
N_crit = K_crit1

fig, ax = plt.subplots(figsize=(12,12))
ax.plot(k_list, N_eq, label="$N^{*}$", color = 'mediumseagreen')
ax.plot(k_list, C_eq, label="$C^{*}$", color = 'mediumvioletred')
ax.plot(k_list, P_eq, label="$P^{*}$", color = 'sandybrown', alpha = 0.7)

crit_line_style = {"alpha":0.4, "linestyle":"--"}
ax.axvline(K_crit1, **crit_line_style, color='mediumvioletred', 
    label="$K_{crit, 1} = \\frac{d_{c}}{\\varepsilon_{c} b_{c}}$")
ax.axvline(K_crit2, **crit_line_style, color='sandybrown', 
    label="$K_{crit, 2} = \\frac{d_{c}}{\\varepsilon_{c} b_{c}} \\left( \\frac{1}{1-\\frac{d_{p}}{\\varepsilon_{p} b_{p}} \\frac{b_{c}}{r}} \\right)$")

crit_line_style = {"alpha":0.4, "linestyle":":"}
ax.axhline(N_crit, **crit_line_style, color = 'mediumseagreen', 
    label="$N^{*} = \\frac{d_{c}}{\\varepsilon_{c} b_{c}}$")
ax.axhline(C_crit, **crit_line_style, color = 'mediumvioletred', 
    label="$C^{*} = \\frac{d_{p}}{\\varepsilon_{p} b_{p}}$")


ax.legend(bbox_to_anchor=(1,1))
ax.grid(alpha = 0.5)

ax.set_xlabel("Carrying capacity K, units $\\frac{kg}{m^{2}}$")
ax.set_ylabel("Producer (N), Consumer (C) and Predator (P) density")
```
