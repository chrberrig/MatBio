```python
# Import stuff: 
from scipy.integrate import odeint, solve_ivp
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from sympy import Symbol
from sympy.solvers import solve

# This is for reasonable fontsize universally defined:
fs_label = 16
parameters = {                
                'figure.titlesize': fs_label+6,
                'axes.labelsize': fs_label,
                'axes.titlesize': fs_label+4,
                'xtick.labelsize': fs_label,
                'ytick.labelsize': fs_label, 
                'legend.fontsize': fs_label, 
                'lines.linewidth': 5
             }
plt.rcParams.update(parameters)
```

```python
# Parameters: 
# TODO: use the real values
params_dict=dict(
    L = 100, 
    N = 1000, 
    D_const = 0.05, 
    v_const = 0, 
    r_const = 0 
)

for k, v in params_dict.items():
    assign_str = f"{k} = {v}"
    exec(assign_str)
    print(assign_str)

ddx = lambda arr, delta: np.array([arr[0]] + 
                                  list((np.diff(arr[:-1])/delta + np.diff(arr[1:])/delta)/2) + 
                                  [arr[-1]])

#def ddx2(arr, delta): 
#    arr[1:-1] = (arr[2:] - 2*arr[1:-1] + arr[:-2])/(delta**2)
#    return arr
    
```

```python
deltax = L/N
x_list = np.arange(0, L, deltax)

#print(x_list)

#D = lambda x, t: np.where([D_const for i in range(N)]==x)#[0]
#v = lambda x, t: np.where([v_const for i in range(N)]==x)#[0]
#r = lambda x, t: np.where([r_const for i in range(N)]==x)#[0]
  
#print(D(5.0, 0))

D = lambda x, t: D_const
v = lambda x, t: v_const
r = lambda x, t: r_const    

J = lambda c, x, t, delta: c*v(x, t) - D(x, t)*ddx(c, delta)

def DAR(state, t, *params):
    
    for k, v in params:
        assign_str = f"{k} = {v}"
        exec(assign_str)
    
    C = state
    dCdt = -ddx(J(C, x_list, t, deltax), deltax) #+ r(x_list, t)*C
    #dCdt = J(C, x_list, t, deltax)
    
    return np.array(dCdt)

C_init = np.zeros(N)
C_init[int(N*1/3):int(N*2/3)] = 1
params = tuple(params_dict.items())

t_list = np.linspace(0, 1000, 10000)#0000)
ns = odeint(DAR, C_init, t_list, args=params)

print(np.shape(ns))
```

```python
fig, ax = plt.subplots(figsize=(16,8))
im = ax.imshow(ns.T, aspect="auto")
ax.set_xlabel("$t$")
ax.set_ylabel("$x$")

```

```python
fig, ax = plt.subplots(figsize=(16,8))
num_levels = 10
levels = [int(k*len(ns)/num_levels) for k in range(num_levels)]
colors = plt.cm.jet(np.linspace(0,1,num_levels))
for i, l in enumerate(levels):
    ax.plot(ns[l], label="C", color=colors[i], alpha=0.2)
#    
#ax.set_yscale("log")
```
